```python
def limpiar_moneda(x):
    if isinstance(x, str):
        return (x.replace('$','').replace(',','').replace('\D',''))
    return (x)
```

```python
valid = '1234567890.' 
def curate(data):
    return float(''.join(filter(lambda char: char in valid, data)))
```

```python
def color_rojo(val):
   color = 'red' if val < -0.7 else 'black'
   return 'color: %s' % color
```

```python
paises = pd.read_csv('economic_freedom_index2019_data.csv', encoding='latin-1')
```

```python
ejem=paises[['2019 Score','Country']]
ejem = ejem.sort_values(by='2019 Score', ascending=False).head(20)
ejem=ejem.set_index('Country')
ax = ejem['2019 Score'].plot(kind='bar')
ax.set_xlabel('Countries', fontsize=12)
ax.set_ylabel('2019 Score', fontsize=12)
plt.show()
```

```python
ejem=paises[['2019 Score','Country']]
ejem = ejem.sort_values(by='2019 Score', ascending=False).tail(20)
ejem=ejem.set_index('Country')
ax = ejem['2019 Score'].plot(kind='bar')
ax.set_xlabel('Countries', fontsize=12)
ax.set_ylabel('2019 Score', fontsize=12)
plt.show()
```
